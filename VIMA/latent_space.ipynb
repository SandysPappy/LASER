```python
# import PCA, TSNE
from sklearn.decomposition import PCA
import numpy as np
import torch
from glob import glob
import matplotlib.pyplot as plt
```

```python
from dataset_builder import get_laser_dataset
```

```python
dataset = get_laser_dataset(task="all", partition="all")
```

```python
len(dataset.attack_embeddings)
```

```python
pca = PCA(n_components=2)
pca.fit(dataset.attack_embeddings)
```

```python
print(pca.explained_variance_ratio_)
```

```python
print(pca.singular_values_)
```

```python
results = pca.transform(dataset.attack_embeddings)
```

```python
# Plot results
plt.scatter(results[:, 0], results[:, 1])
8plt.show()
```
